```python
import json
```

```python

# Load the JSON file with 'utf-8' encoding
with open('D:/work_place/data_example.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
    
print("File loaded successfully.")
```

```python

# Function to structure and clean nested JSON (but preserve null and empty values)
def structure_data(item):
    # Parsing 'data' field if it's a string (nested JSON)
    if 'data' in item and isinstance(item['data'], str):
        item['data'] = json.loads(item['data'])  # Parse the nested JSON string
    return item

# Structure and clean the entire dataset (without removing nulls or empty values)
structured_data = [structure_data(item) for item in data]
```

```python

# Optionally, you can sort the data by a specific key (e.g., 'name') if desired
def sort_data(data, key):
    return sorted(data, key=lambda x: x.get(key, ""))  # Sort by the provided key, defaulting to empty string

# Example: Sort by 'name' (optional step, can be skipped if not needed)
sorted_structured_data = sort_data(structured_data, 'name')
```

```python

# Save the structured and cleaned data back to a new file
with open('D:/work_place/structured_data_example.json', 'w', encoding='utf-8') as file:
    json.dump(sorted_structured_data, file, indent=4, ensure_ascii=False)

print("Data structured and saved successfully.")
```

```python
# Function to count the number of fields in each row (dictionary)
def count_fields_per_row(data):
    counts = []
    for index, row in enumerate(data):
        field_count = len(row)  # Counting the number of keys in each row (dictionary)
        counts.append((index, field_count))
    return counts

# Get the count of fields for each row
field_counts = count_fields_per_row(data)

# Print the number of fields in each row
for index, count in field_counts:
    print(f"Row {index + 1}: {count} fields")
```

```python

# Function to recursively extract unique field names under each dictionary
def find_fields_per_dictionary(data, parent="root", fields_per_dict=None):
    if fields_per_dict is None:
        fields_per_dict = {}  # Dictionary to store fields for each unique dictionary

    # If current data is a dictionary, we iterate over its items
    if isinstance(data, dict):
        if parent not in fields_per_dict:
            fields_per_dict[parent] = set()  # Create a set for storing fields of this dictionary

        # Add the keys (field names) to the set
        for key, value in data.items():
            fields_per_dict[parent].add(key)

            # Recursively call the function for nested dictionaries or lists
            if isinstance(value, dict):
                find_fields_per_dictionary(value, parent=key, fields_per_dict=fields_per_dict)
            elif isinstance(value, list):
                # For lists, process each item
                for idx, item in enumerate(value):
                    find_fields_per_dictionary(item, parent=f"{key}[{idx}]", fields_per_dict=fields_per_dict)

    # If current data is a list, process each item
    elif isinstance(data, list):
        for idx, item in enumerate(data):
            find_fields_per_dictionary(item, parent=f"{parent}[{idx}]", fields_per_dict=fields_per_dict)

    return fields_per_dict

# Get the fields for each unique dictionary in the JSON data
fields_per_dict = find_fields_per_dictionary(data)
```

```python

# Print the fields under each unique dictionary
for dict_name, fields in fields_per_dict.items():
    print(f"\nDictionary: {dict_name}")
    print("Fields:")
    for field in sorted(fields):
        print(f"  - {field}")
```

```python

# Recursive function to categorize and structure keys based on their type (dict, list, or simple value)
def categorize_keys(data, parent="root", level=0):
    result = []  # List to store the categorized structure

    # If the current element is a dictionary
    if isinstance(data, dict):
        for key, value in data.items():
            prefix = "  " * level  # Indentation based on the level of nesting
            # If value is another dictionary, recurse
            if isinstance(value, dict):
                result.append(f"{prefix}{key} (dictionary):")
                result.extend(categorize_keys(value, parent=key, level=level + 1))
            # If value is a list, process the list
            elif isinstance(value, list):
                result.append(f"{prefix}{key} (list):")
                result.extend(categorize_list(value, parent=key, level=level + 1))
            else:
                # If it's a simple value (str, int, float, bool, etc.), add its type
                result.append(f"{prefix}{key} ({type(value).__name__})")

    return result

# Function to categorize the elements of a list
def categorize_list(lst, parent="root", level=0):
    list_structure = []
    prefix = "  " * level  # Indentation based on the level of nesting

    for idx, item in enumerate(lst):
        # If the item is a dictionary, recurse into the dictionary
        if isinstance(item, dict):
            list_structure.append(f"{prefix}Item {idx} (dictionary):")
            list_structure.extend(categorize_keys(item, parent=f"{parent}[{idx}]", level=level + 1))
        # If the item is another list, recursively categorize it
        elif isinstance(item, list):
            list_structure.append(f"{prefix}Item {idx} (list):")
            list_structure.extend(categorize_list(item, parent=f"{parent}[{idx}]", level=level + 1))
        else:
            # If it's a simple value, append its type
            list_structure.append(f"{prefix}Item {idx} ({type(item).__name__})")

    return list_structure

# Call the function to map and categorize all fields in the JSON
categorized_structure = categorize_keys(data)

# Print the categorized structure as a readable list
for line in categorized_structure:
    print(line)
```
